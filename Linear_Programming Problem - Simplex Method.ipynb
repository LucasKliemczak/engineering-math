## **Linear Programming Problem - Simplex Method**
$$
Min \ Z = 5x_1 + 7x_2 +12x_3
$$
$$
x_1 + x_2 + x_3 \geq 500
$$
$$
x_1 - 2x_2 \geq 0
$$
$$
x_3 \ge 340
$$
$$
x_1,x_2,x_3 \ge 0
$$

```python
import numpy as np
from scipy.optimize import linprog

Z = np.array([5,7,12]) # objective function

# inequality constraint

A = np.array([[-1,-1,-1],
              [-1, 2, 0],
               [0,0,-1]])

B = np.array([-500,0,-340])

x_low = 0
x_upper = None

result = linprog(Z,A,B,method='revised simplex',bounds=[x_low,x_upper])

print('Minimum Value of Objective Function', result.fun)
print('Values of x1,x2,x3: ', result.x)
```

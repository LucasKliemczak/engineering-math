## **Newton Raphson Numerical Method**

$$
x_n+_1 = x_n - \frac{f(x_n)}{f'(n_n)}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

def f(x):
  return x**3/2 - 2*x -1

def g(x):
  return derivative(f,x,dx=1e-4)

x0 = 1
x1=1
error = 1e-5
vector = []
while abs(f(x0))>error:
  x1 = x0 - (f(x0)/g(x0))
  x0 = x1
  vector.append(x1)
print(vector)   # concertar isso. Descobrir como colocar os pontos no gráfico


plt.grid(linestyle='--', color='black')
```
